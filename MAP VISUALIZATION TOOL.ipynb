```python
import pandas as pd
df = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/Police_Department_Incidents_-_Previous_Year__2016_.csv')
```

```python
df.shape
```

```python
limit = 100
df = df.iloc[0:limit, :]
```

```python
df.shape
```

```python
df.head()
```

```python
sanfransisco = folium.Map(location=[37.77, -122.42], zoom_start = 12)
sanfransisco
```

```python
incidents = folium.map.FeatureGroup()
for lat, lng in zip(df.Y, df.X):
    incidents.add_child(
        folium.features.CircleMarker(  [37.77, -122.42],
                                    radius=10, color='yellow', fill =True, fill_color ='blue', fill_opacity = 0.6)
          )
```

```python
sanfransisco.add_child(incidents)
```

```python
latitudes = list(df.Y)
longitudes = (df.X)
labels = list(df.Category)
```

```python
for lat, lng, label in zip(latitudes, longitudes, labels):folium.Marker([37.77, -122.42], popup = label).add_to(sanfransisco)
sanfransisco.add_child(incidents)
```

```python
from folium import plugins
incidents = plugins.MarkerCluster().add_to(sanfransisco)
for lat, lng, label in zip(df.Y, df.X, df.Category):
    folium.Marker(location =[37.77, -122.42], icon = None, 
    popup=label,
    ).add_to(incidents)
```
